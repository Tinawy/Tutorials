```python
import numpy as np
hist = np.zeros([3,256])
for c in range(3):
    for i in range(im.shape[0]):
        for j in range(im.shape[1]):
            hist[c,im[i,j,c]] += 1
# print(his)
```

```python
mpl.figure(figsize=[15,5])
mpl.subplot(1,3,1)
mpl.plot(range(0,256),hist[0])
mpl.subplot(1,3,2)
mpl.plot(range(0,256),hist[1])
mpl.subplot(1,3,3)
mpl.plot(range(0,256),hist[2])
```

```python
print('Features')
print('R')
print(np.mean(hist[0]), np.std(hist[0]), np.median(hist[0]))
print('G')
print(np.mean(hist[1]), np.std(hist[1]), np.median(hist[1]))
print('B')
print(np.mean(hist[2]), np.std(hist[2]), np.median(hist[2]))
```
